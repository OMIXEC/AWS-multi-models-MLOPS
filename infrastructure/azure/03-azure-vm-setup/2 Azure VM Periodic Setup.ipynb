# Azure VM Periodic Setup
Start, stop, and deallocate Azure VMs.


```python
from azure.identity import DefaultAzureCredential
from azure.mgmt.compute import ComputeManagementClient

credential = DefaultAzureCredential()
subscription_id = 'your-subscription-id'
compute_client = ComputeManagementClient(credential, subscription_id)

resource_group_name = 'mlops-rg'
vm_name = 'mlops-prod-vm'
```

```python
def start_vm():
    print('Starting VM...')
    compute_client.virtual_machines.begin_start(resource_group_name, vm_name)
    print('Started.')

start_vm()
```

```python
def stop_vm():
    print('Deallocating VM...')
    compute_client.virtual_machines.begin_deallocate(resource_group_name, vm_name)
    print('Stopped.')

# stop_vm()
```
