# GCP Compute Engine Periodic Setup
Start and stop Compute Engine instances.


```python
from google.cloud import compute_v1
import time

project_id = 'your-gcp-project-id'
zone = 'us-central1-a'
instance_name = 'mlops-prod'

client = compute_v1.InstancesClient()
```

```python
def start_instance():
    print('Starting instance...')
    operation = client.start(project=project_id, zone=zone, instance=instance_name)
    print('Started.')

start_instance()
```

```python
def stop_instance():
    print('Stopping instance...')
    operation = client.stop(project=project_id, zone=zone, instance=instance_name)
    print('Stopped.')

# stop_instance()
```
