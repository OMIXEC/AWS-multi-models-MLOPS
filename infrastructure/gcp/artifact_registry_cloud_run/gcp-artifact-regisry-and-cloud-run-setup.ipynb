## 1. Upload Model Artifact to Google Cloud Storage (GCS)
Ensure your model artifacts are stored in GCS so the container can download them at runtime.

```python
from google.cloud import storage
client = storage.Client()
bucket_name = 'your-ml-models-bucket-gcp'
def upload_model():
    bucket = client.bucket(bucket_name)
    blob = bucket.blob('models/model.pt')
    blob.upload_from_filename('model.pt')
    print('Model uploaded to GCS.')

# upload_model()
```

## 2. Push to Google Artifact Registry (GAR)

```python
# Build and push the image using gcloud and docker
!gcloud auth configure-docker us-central1-docker.pkg.dev
!docker build -t mlops-api .
!docker tag mlops-api us-central1-docker.pkg.dev/YOUR_PROJECT_ID/YOUR_REPO_NAME/mlops-api:latest
!docker push us-central1-docker.pkg.dev/YOUR_PROJECT_ID/YOUR_REPO_NAME/mlops-api:latest
```

## 3. Deploy to Cloud Run

```python
!gcloud run deploy mlops-api \\
    --image us-central1-docker.pkg.dev/YOUR_PROJECT_ID/YOUR_REPO_NAME/mlops-api:latest \\
    --region us-central1 \\
    --platform managed \\
    --allow-unauthenticated \\
    --port 80
```
